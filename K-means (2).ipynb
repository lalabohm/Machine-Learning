# Método de  agrupamento não supervisionado usando K-means

### Os clusters são grupos e podem ser representados num gráfico orientado por xy, por um mapa de calor (heatmap), etc.
#### 0 "k" representa a quantidade de clusters.

#### Passo 1: defina o valor de K. (quantos clusters serão definidos).
#### Passo 2: escolha K pontos, aleatoriamente, no gráfico.
#### Passo 3: meça a distância de um ponto para o cada cluster.
#### Passo 4: atribua o ponto ao cluster que ele estiver mais perto. (repita com os outros pontos)
#### Passo 5: faça a média dos clusters. Meça a distância entre os clusters.

### Importação de bibliotecas

```python
from sklearn.cluster import KMeans #focada em algoritmos de agrupamento
import numpy as np
import matplotlib.pyplot as plt
```

### Geração de dados

```python
np.random.seed(42) #determina a sequencia especifica de numeros aleatorios que sera gerada pelo algoritmo
data = np.random.randn(300, 2) #matriz de 300 linhas e 2 colunas 
data[:100] += 5, 5 #coordenadas dos 3 grupos
data[100:200] += 15, 5
data[200:] += 10, 10
```

### Aplicação do K-means

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data) #calculando a distância de cada ponto a todos os centroides,depois atualiza o novo centroide com o ponto médio de todos os pontos atualmente atribuídos a esse cluster 
labels = kmeans.labels_ #obtém os rotulos dos clusters atribuidos a cada ponto
```

### Visualização dos Resultados

```python
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o')
centers = kmeans.cluster_centers_ #obtem as coordenadas dos centros dos clusters.
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.title('Exemplo de K-means clustering')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```
